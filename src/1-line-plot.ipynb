### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = x**2

# Error handling function
def validate_line_data(x, y):
    """Checks if x and y are valid for a line plot."""
    if len(x) == 0 or len(y) == 0:
        raise ValueError("Error: x and y values cannot be empty.")
    if len(x) != len(y):
        raise ValueError("Error: x and y must have the same length.")
    if any(not isinstance(i, (int, float)) for i in x) or any(not isinstance(i, (int, float)) for i in y):
        raise TypeError("Error: x and y values must be integers or floats.")

# Validate the data
validate_line_data(x, y)

# Create plot
plt.plot(x, y)

# Customize the plot
plt.title('Plot of y = x^2')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)

# Show the plot
plt.show()
```
