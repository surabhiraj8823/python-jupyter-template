### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Error handling function
def validate_bar_data(categories, values):
    """Checks if categories and values are valid for a bar chart."""
    if not categories or not values:
        raise ValueError("Error: Categories and values cannot be empty.")
    if len(categories) != len(values):
        raise ValueError("Error: Categories and values must have the same length.")
    if any(not isinstance(value, (int, float)) for value in values):
        raise TypeError("Error: Values must be integers or floats.")

# Validate the data
validate_bar_data(categories, values)

# Create the bar plot
plt.bar(categories, values, color=colors)

# Add title and labels
plt.title("Category-wise Values")
plt.xlabel("Categories")
plt.ylabel("Values")

# Show the plot
plt.show()
```
