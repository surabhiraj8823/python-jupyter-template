### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Error handling function
def validate_scatter_data(x, y):
    """Checks if x and y are valid for a scatter plot."""
    if not x or not y:
        raise ValueError("Error: X and Y data cannot be empty.")
    if len(x) != len(y):
        raise ValueError("Error: X and Y must have the same length.")
    if any(not isinstance(i, (int, float)) for i in x + y):
        raise TypeError("Error: X and Y values must be integers or floats.")

# Validate the data
validate_scatter_data(x, y)

# Create the scatter plot
plt.scatter(x, y, color='blue', label='Data Points')

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='right')

# Add title and labels
plt.title("Scatter Plot with Annotations")
plt.xlabel("X values")
plt.ylabel("Y values")

# Show the plot
plt.legend()
plt.grid(True)
plt.show()
```
