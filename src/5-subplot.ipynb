### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Error handling function
def validate_data(*args):
    """Checks if provided datasets are valid (not empty, correct types, and consistent)."""
    for data in args:
        if not data:
            raise ValueError("Error: One of the data lists is empty. Please provide valid input.")
        if isinstance(data, list) and any(not isinstance(i, (int, float)) for i in data if isinstance(i, (int, float))):
            raise TypeError("Error: All numerical values must be integers or floats.")

# Validate numerical data
validate_data(x, y, values, sizes)

# Validate categorical data
if len(categories) != len(values):
    raise ValueError("Error: Categories and values must have the same length.")

if len(labels) != len(sizes):
    raise ValueError("Error: Labels and sizes for the pie chart must have the same length.")

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line chart (Top-left)
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='blue')
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X values")
axs[0, 0].set_ylabel("Y values")

# Bar chart (Top-right)
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Scatter plot (Bottom-left)
axs[1, 0].scatter(x, y, color='orange', label='Data Points')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X values")
axs[1, 0].set_ylabel("Y values")
axs[1, 0].legend()
axs[1, 0].grid(True)

# Pie chart (Bottom-right)
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'orange'], startangle=140, shadow=True)
axs[1, 1].set_title("Pie Chart")

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()
```
